```python
## Executive Summary

"""The analysis was conducted on 10,523 cleaned price observations collected from two Moroccan e-commerce websites: Jumia Morocco and Electroplanet. The dataset covers 934 unique products, 18 product categories, and the period from May 1, 2026 to June 10, 2026.

The main objective was to analyze price behavior using descriptive and inferential statistics. The results show that product prices are highly heterogeneous, strongly influenced by product category, and significantly different between the two websites in the observed dataset."""
```

```python
## Key Dataset Facts

"""- Total cleaned observations: 10,523
- Unique products: 934
- Websites: 2
- Product categories: 18
- Observation period: May 1, 2026 to June 10, 2026
- Main analytical tables used: `clean_prices`, `price_time_series`, and `price_intelligence`"""
```

```python
## Descriptive Insights

"""The global average product price is approximately 1,801.67 MAD, while the median price is 979 MAD. The mean is much higher than the median, indicating a right-skewed price distribution caused by expensive products such as gaming PCs, MacBooks, and ultrabooks.

The standard deviation is approximately 2,971.10 MAD, showing strong price dispersion across the catalog. Prices range from 15 MAD to 39,999 MAD, confirming that the dataset includes both low-cost accessories and high-end electronics.

At the website level, Electroplanet shows a higher average and median price than Jumia Morocco. However, this comparison should be interpreted carefully because Jumia has many more observations and includes a broader range of low-cost products.

At the category level, premium computer-related categories such as `pc-gamer`, `ultrabook`, `macbook`, `pc-hybride`, and `notebook` have the highest price levels. In contrast, accessory-related categories have much lower prices."""
```

```python
## Inferential Insights

"""Normality tests showed that price distributions are not normally distributed, even after log transformation. Therefore, non-parametric tests were used for group comparisons.

The Mann-Whitney U test confirmed a statistically significant difference between price distributions on Jumia Morocco and Electroplanet. Electroplanet prices are higher in the observed dataset, but this result is partly influenced by catalog composition.

The Kruskal-Wallis test confirmed that price distributions differ significantly across product categories. Post-hoc pairwise tests showed that most category pairs are significantly different, especially comparisons between premium computer categories and low-cost accessory categories.

Correlation analysis showed a weak positive relationship between price and rating, and a moderate negative relationship between price and review count. This suggests that higher-priced products tend to have slightly higher ratings, while products with more reviews tend to be lower-priced.

The regression analysis showed that rating, review count, and time explain approximately 29% of the variation in log-prices. When product category is added to the model, the explained variance increases to approximately 70%, confirming that category is the strongest driver of price differences."""
```

```python
## Confidence Intervals and Power Analysis

"""Bootstrap confidence intervals were used because price distributions are skewed and contain extreme values. The confidence intervals showed that Electroplanet has a substantially higher estimated mean price than Jumia Morocco, and the intervals for the two websites do not overlap.

Power analysis showed that the observed effect size between Jumia and Electroplanet log-prices is very large. The available sample sizes are sufficient to detect the observed difference with adequate statistical power."""
```

```python
## Limitations

"""The analysis has several limitations. First, the dataset is imbalanced between websites, with many more observations from Jumia Morocco than from Electroplanet. Second, the product catalogs are not identical across platforms, so platform-level price differences may reflect product mix rather than pure pricing strategy.

Third, `rating` and `review_count` contain many missing values, mainly due to scraping limitations and product pages without available customer feedback. Analyses involving these variables were therefore performed only on complete cases.

Finally, the observation period covers approximately six weeks, which is sufficient for short-term price monitoring but limited for detecting long-term seasonality."""
```

```python
## Final Conclusion

"""The analysis confirms that e-commerce price behavior is strongly influenced by product category, website, and customer engagement variables. While Electroplanet appears more expensive than Jumia Morocco in the observed dataset, the difference should be interpreted in the context of catalog composition.

The most important analytical finding is that product category explains a large part of price variation. Therefore, future price intelligence dashboards and alerts should prioritize category-level monitoring, product-level trends, and volatility indicators rather than relying only on global average prices.

Overall, the project successfully demonstrates how transformed e-commerce data can be used for descriptive statistics, inferential testing, regression modeling, confidence intervals, and business-oriented price intelligence."""
```
